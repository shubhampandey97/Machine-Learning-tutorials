#### MICE Imputation Using Loop (Manual Version)

```python
import pandas as pd
import numpy as np

df = np.round(pd.read_csv(r'D:/Data Science/Machine Learning/Data/50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
df.head()
```

```python
np.random.seed(42)

df.loc[df.sample(5).index, "R&D Spend"] = np.nan
df.loc[df.sample(5).index, "Marketing Spend"] = np.nan
df.loc[df.sample(3).index, "Administration"] = np.nan
```

```python
df.isnull().sum()
```

```python
X = df.drop("Profit", axis=1)
y = df["Profit"]
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=2
)
```

```python
print("Missing values in X_train:\n")
print(X_train.isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# -------------------------------
# Copy training data
# -------------------------------
X_train_mice = X_train.copy()

print("\n📌 Original Training Data (with missing values):")
display(X_train_mice.head(10))

# -------------------------------
# Step 1: Initial Mean Imputation
# -------------------------------
init_imputer = SimpleImputer(strategy="mean")
X_train_mice[:] = init_imputer.fit_transform(X_train_mice)

print("\n✅ After Initial Mean Imputation:")
display(X_train_mice.head(10))

# -------------------------------
# Columns with missing values
# -------------------------------
missing_cols = X_train.columns[X_train.isnull().any()]
print("\nColumns with missing values:", missing_cols)

# -------------------------------
# Step 2: Iterative MICE Loop
# -------------------------------
n_iterations = 5

for iteration in range(n_iterations):

    print(f"\n==============================")
    print(f"   ITERATION {iteration+1}/{n_iterations}")
    print(f"==============================")

    for col in missing_cols:

        # Find indices where original training data had missing values
        missing_index = X_train[X_train[col].isnull()].index

        if len(missing_index) == 0:
            continue

        # Split dataset
        train_data = X_train_mice.drop(index=missing_index)
        test_data  = X_train_mice.loc[missing_index]

        X_col_train = train_data.drop(columns=[col])
        y_col_train = train_data[col]

        X_col_test = test_data.drop(columns=[col])

        # Train regression model
        model = LinearRegression()
        model.fit(X_col_train, y_col_train)

        # Predict missing values
        predicted_values = model.predict(X_col_test)

        # Fill values back
        X_train_mice.loc[missing_index, col] = predicted_values

    # ✅ Show dataframe after each iteration
    print(f"\n📌 Data After Iteration {iteration+1}:")
    display(X_train_mice.head(10))

print("\n🎉 Final MICE Imputation Completed!")
```

```python
X_test_mice = init_imputer.transform(X_test)
```

```python
print(pd.DataFrame(X_train_mice).isnull().sum())
```

#### True MICE in Sklearn

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

mice = IterativeImputer(max_iter=10, random_state=42)

X_train_mice = mice.fit_transform(X_train)
X_test_mice  = mice.transform(X_test)
print(pd.DataFrame(X_train_mice).isnull().sum())
```
