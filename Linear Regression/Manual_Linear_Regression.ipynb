# Simple Linear Regression

```python
class LinearRegressionModel:
    def __init__(self):
        self.m = None
        self.c = None

    def fit(self, X_train, y_train):
        # print(X_train.shape, y_train.shape)
        numerator = 0
        denominator = 0

        for i in range(len(X_train)):
            numerator = numerator + ((X_train[i]-X_train.mean())*(y_train[i]-y_train.mean()))
            denominator = denominator + ((X_train[i]-X_train.mean())*(X_train[i]-X_train.mean()))

        self.m = numerator/denominator
        self.c = y_train.mean() - (self.m * X_train.mean())
        print(f"Coefficient: {self.m}, Intercept: {self.c}")

    def predict(self, X_test):
        y_predicted = self.m * X_test + self.c
        print(f"Predicted values: {y_predicted}")
```

# Multiple Linear Regression

```python
class MultipleLinearRegressionModel:
    def __init__(self):
        self.coeficient_ = None
        self.intercept_ = None

    def fit(self, X_train, y_train):
        X_train = np.insert(X_train, 0, 1, axis=1)


        # self.coeficient_ = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
        betas = np.linalg.inv(np.dot(X_train.T, X_train)).dot(X_train.T).dot(y_train)
        self.intercept_ = betas[0]
        self.coeficient_ = betas[1:]

    def predict(self, X_test):
        y_predicted = np.dot(X_test, self.coeficient_) + self.intercept_
        return y_predicted
```
